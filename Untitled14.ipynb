#**4.Uniendo los datos**

##**Población de los estados**

```python
poblacion_estados = pd.read_excel('POP2022_Municipios.xls',header = 1, skipfooter = 34)
poblacion_estados
```

```python
poblacion_estados.groupby('UF').sum(numeric_only=True)
```

```python
# poblacion_estados['POPULAÇÃO'].astype(int)
```

```python
poblacion_estados[poblacion_estados['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
poblacion_estados = poblacion_estados.assign(poblacion_sin_parentesis = poblacion_estados['POPULAÇÃO'].replace('\(\d{1,2}\)','',regex=True),
                         poblacion = lambda x: x.loc[:,'poblacion_sin_parentesis'].replace('\.','',regex=True))
```

```python
poblacion_estados[poblacion_estados['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
poblacion_estados = poblacion_estados.astype({'poblacion':'int64'})
```

```python
poblacion_estados = poblacion_estados.groupby('UF')[['poblacion']].sum().reset_index()
poblacion_estados
```

```python
emisiones_estados = emisiones_por_año[emisiones_por_año['Año'] == 2021].groupby('Estado')[['Emision']].sum(numeric_only=True).reset_index()
emisiones_estados
```

```python
datos_agrupados = pd.merge(emisiones_estados, poblacion_estados,left_on='Estado',right_on='UF')
datos_agrupados
```

```python
datos_agrupados.plot(x = 'poblacion', y = 'Emision', kind = 'scatter', figsize = (8,6));
```

```python
import plotly.express as px
```

```python
px.scatter(datos_agrupados, x = 'poblacion', y = 'Emision', hover_name = 'Estado')
```

```python
px.scatter(datos_agrupados, x = 'poblacion', y = 'Emision', text = 'Estado', opacity = 0)
```

```python
datos_agrupados = datos_agrupados.assign(emision_per_capital = datos_agrupados['Emision'] / datos_agrupados['poblacion']).sort_values('emision_per_capital', ascending=False)
datos_agrupados
```

```python
px.bar(datos_agrupados, x = 'Estado', y = 'emision_per_capital')
```

```python
px.scatter(data_frame = datos_agrupados, x = 'poblacion', y = 'Emision', text = 'Estado', size = 'emision_per_capital')
```
